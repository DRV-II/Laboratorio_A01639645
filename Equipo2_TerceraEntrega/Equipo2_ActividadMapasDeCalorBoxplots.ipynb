```python
# Primero, importaremos las librerías básicas que usaremos en los siguientes
# ejemplos.
import numpy as np
import pandas as pd

# Importamos la librería.
import matplotlib.pyplot as plt

# la forma general de graficar con `seaborn es:

# Importamos la librería
import seaborn as sns
# seaborn hace más bonitas las gráficas.
# Usamos set para definir el estilo de Seaborn para todas las gráficas.

```

### Lectura del archivo csv dentro de la variable df

```python
df = pd.read_csv("Beneficiarios_emb_mayores.csv")
df.info()
df.head()
```

### Se creo un gráfico en donde se ploteo el monto de apoyo por cada Estado

```python
# Configuramos el tamaño de imagen
fig = plt.figure(figsize=(6,4))

# Graficamos los incresos contra el límite de crédito
plt.plot(df['CLAVE ESTADO'], df['MONTO CONAPESCA'], '*')

# Agregamos títulos a los ejes y al gráfico
plt.xlabel('CLAVE ESTADO')
plt.ylabel('MONTO CONAPESCA')
plt.title('Gráfico de un conjunto de datos')

# Aquí la leyenda hace mucho más sentido
plt.legend(loc='best')

# Agregamos la cuadrícula para que se vea mejor
plt.grid(True)
```

### Gráfica en donde se mide la frecuencia de cantidad del monto conapesca

```python
# Histograma simple
sns.histplot(data=df, x='MONTO CONAPESCA')
```

### Ploteo del Heatmap para visualizar que datos tienen más correlación 

```python
# Vamos a graficar la matriz de correlación del dataset Iris
monto_corr = df.corr()

# Gráfico heatmap. Seleccionamos los valores extremos con vmin y vmax.
# El mapa de color que usaremos es de un extremo azul y del otro rojo.
# Con annot podemos desplegar el valor de cada celda
# Con square hacemos que el gráfico sea simétrico en tamaño de ejes
sns.heatmap(data=monto_corr, vmin=-1, vmax=1, cmap = 'RdBu', annot=True, square = True)
```

### Gráfico boxplot de CLAVE ESTADO

```python


# Tamaño de la imagen
fig = plt.figure(figsize=(9, 6))

# Gráfico boxplot
sns.boxplot(data=df, x='CLAVE ESTADO')

# Ejes y título
plt.title('Histograma de CLAVE ESTADO')


```

### Gráfico en donde se visualiza la comparación de distribución del MONTO CONAPESA junto la CLAVE ESTADO

```python
# Tamaño de la imagen
fig = plt.figure(figsize=(9,6))

# Gráfico
sns.histplot(data=df, x='MONTO CONAPESCA', hue='CLAVE ESTADO', bins=20, kde=True)

# Ejes y título
plt.xlabel('MONTO CONAPESCA')
plt.ylabel('Frecuencia')
plt.title('Comparación de la distribución del MONTO CONAPESCA por CLAVE ESTADO')
```

### Creación de gráfica de distribución donde se visualiza la relación entre CLAVE MUNICIPIO y ANIO APOYO por CLAVE ESTADO

```python
# Graficaremos la relación entre el ancho y la longitud del sépalo.

# Tamaño de la imagen.
fig = plt.figure(figsize=(6, 4))

# Gráfico scatterplot.
sns.scatterplot(data=df, x ='CLAVE MUNICIPIO', y='ANIO APOYO', hue='CLAVE ESTADO')

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Relación entre CLAVE MUNICIPIO y ANIO APOYO por CLAVE ESTADO')
plt.xlabel('CLAVE MUNICIPIO')
plt.ylabel('ANIO APOYO')
```

### Gráfica en donde se observa la frecuencia en la que aparece cada estado por su CLAVE ESTADO

```python
# Vamos a graficar el conteo de la variable categórica (Species)
# Tamaño de la imagen
fig = plt.figure(figsize=(15, 8))

# Gráfico countplot para hacer barras con el número de apariciones de cada especie.
sns.countplot(data=df, y = 'CLAVE MUNICIPIO')

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Observaciones de cada MUNICIPIO')
plt.xlabel('Frecuencia')
plt.ylabel('CLAVE MUNICIPIO')
```

### Gráfica donde se hace la relación entre CLAVE MUNICIPIO y el MONTO CONAPESCA

```python
# Graficaremos la relación entre el ancho y la longitud del sépalo.

# Tamaño de la imagen.
fig = plt.figure(figsize=(15, 8))

# Gráfico scatterplot.
sns.scatterplot(data=df, x ='MONTO CONAPESCA', y='CLAVE MUNICIPIO', hue='CLAVE ESTADO')

# Ejes y título. Colocamos la etiqueta correcta de acuerdo a la orientación.
plt.title('Relación entre CLAVE MUNICIPIO y MONTO CONAPESCA por CLAVE ESTADO')
plt.xlabel('MONTO CONAPESCA')
plt.ylabel('CLAVE MUNICIPIO')
```
